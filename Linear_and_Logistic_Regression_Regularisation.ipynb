Load the breast cancer dataset, split it into training and testing sets, scale the features, train a Logistic Regression model, and evaluate its performance. Additionally, load the diabetes dataset, split it into training and testing sets, scale the features, train Linear, Lasso (with alpha=1), and Ridge (with alpha=1) regression models, and evaluate their performance using Mean Squared Error and model coefficients.

## Load Breast Cancer Dataset

### Subtask:
Load the breast cancer dataset from `sklearn.datasets.load_breast_cancer()`.


**Reasoning**:
Load the breast cancer dataset from scikit-learn, extract features and target into pandas DataFrame and Series respectively.



```python
import pandas as pd
from sklearn.datasets import load_breast_cancer

# 2. Call load_breast_cancer() and store the returned object
cancer_data = load_breast_cancer()

# 3. Extract the feature data (X) and store it in a pandas DataFrame
X = pd.DataFrame(cancer_data.data, columns=cancer_data.feature_names)

# 4. Extract the target data (y) and store it in a pandas Series
y = pd.Series(cancer_data.target, name='target')

print("Features (X) head:")
print(X.head())
print("\nTarget (y) head:")
print(y.head())
```

## Split Breast Cancer Data

### Subtask:
Split the breast cancer dataset into 80:20 training and testing sets using `train_test_split` with `random_state=9001`.


**Reasoning**:
The subtask requires splitting the dataset into training and testing sets. This can be achieved using the `train_test_split` function from `sklearn.model_selection`. The `X` and `y` variables from the previous step are the features and target respectively, which will be used for splitting. Printing the shapes will confirm the split happened correctly.



```python
from sklearn.model_selection import train_test_split

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=9001)

print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

## Scale Breast Cancer Data

### Subtask:
Scale the features of the breast cancer dataset using `MinMaxScaler`.


**Reasoning**:
The subtask requires scaling the features of the breast cancer dataset using `MinMaxScaler`. This involves importing the scaler, instantiating it, fitting it to the training data, and then transforming both the training and testing sets. I will print the first few rows of the scaled dataframes to verify the scaling operation.



```python
from sklearn.preprocessing import MinMaxScaler

# Instantiate a MinMaxScaler object
scaler = MinMaxScaler()

# Fit the scaler to the training data
scaler.fit(X_train)

# Transform both the training and testing data
X_train_scaled = pd.DataFrame(scaler.transform(X_train), columns=X_train.columns)
X_test_scaled = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)

print("X_train_scaled head:")
print(X_train_scaled.head())
print("\nX_test_scaled head:")
print(X_test_scaled.head())
```

## Train Logistic Regression Model

### Subtask:
Train a `LogisticRegression` model (using the default solver) on the scaled breast cancer training data.


**Reasoning**:
The subtask requires training a Logistic Regression model. This involves importing the `LogisticRegression` class, instantiating it, and then fitting it to the scaled training features (`X_train_scaled`) and target (`y_train`).



```python
from sklearn.linear_model import LogisticRegression

# Instantiate a LogisticRegression model object
model = LogisticRegression()

# Fit the LogisticRegression model to the scaled training data
model.fit(X_train_scaled, y_train)

print("Logistic Regression model trained successfully.")
```

## Evaluate Logistic Regression Model

### Subtask:
Evaluate the performance of the trained `LogisticRegression` model on the scaled breast cancer test data using accuracy, precision, recall, and F1-score.

**Reasoning**:
The subtask is to evaluate the performance of the trained Logistic Regression model. This involves importing necessary metrics, making predictions on the scaled test data, and then calculating accuracy, precision, recall, and F1-score.



```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Make predictions on the scaled test data
y_pred = model.predict(X_test_scaled)

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
```

## Load Diabetes Dataset

### Subtask:
Load the diabetes dataset from `sklearn.datasets.load_diabetes()`.

**Reasoning**:
The subtask requires loading the diabetes dataset. This involves importing `load_diabetes` from `sklearn.datasets`, calling the function, and then storing the features and target into pandas DataFrames/Series.



```python
import pandas as pd
from sklearn.datasets import load_diabetes

# Call load_diabetes() and store the returned object
diabetes_data = load_diabetes()

# Extract the feature data (X) and store it in a pandas DataFrame
X_diabetes = pd.DataFrame(diabetes_data.data, columns=diabetes_data.feature_names)

# Extract the target data (y) and store it in a pandas Series
y_diabetes = pd.Series(diabetes_data.target, name='target')

print("Diabetes Features (X_diabetes) head:")
print(X_diabetes.head())
print("\nDiabetes Target (y_diabetes) head:")
print(y_diabetes.head())
```

## Split Diabetes Data

### Subtask:
Split the diabetes dataset into 80:20 training and testing sets using `train_test_split` with `random_state=42`.

**Reasoning**:
The subtask requires splitting the diabetes dataset into training and testing sets. This can be achieved using the `train_test_split` function from `sklearn.model_selection` on the `X_diabetes` and `y_diabetes` variables. Printing the shapes will confirm the split happened correctly.



```python
from sklearn.model_selection import train_test_split

# Split the diabetes dataset into training and testing sets
X_diabetes_train, X_diabetes_test, y_diabetes_train, y_diabetes_test = train_test_split(X_diabetes, y_diabetes, test_size=0.2, random_state=42)

print("Shape of X_diabetes_train:", X_diabetes_train.shape)
print("Shape of X_diabetes_test:", X_diabetes_test.shape)
print("Shape of y_diabetes_train:", y_diabetes_train.shape)
print("Shape of y_diabetes_test:", y_diabetes_test.shape)
```

## Scale Diabetes Data

### Subtask:
Scale the features of the diabetes dataset using `MinMaxScaler`.

**Reasoning**:
The subtask requires scaling the features of the diabetes dataset using `MinMaxScaler`. This involves importing the scaler, instantiating it, fitting it to the training data, and then transforming both the training and testing sets. I will print the first few rows of the scaled dataframes to verify the scaling operation.



```python
from sklearn.preprocessing import MinMaxScaler

# Instantiate a MinMaxScaler object
scaler_diabetes = MinMaxScaler()

# Fit the scaler to the training data
scaler_diabetes.fit(X_diabetes_train)

# Transform both the training and testing data
X_diabetes_train_scaled = pd.DataFrame(scaler_diabetes.transform(X_diabetes_train), columns=X_diabetes_train.columns)
X_diabetes_test_scaled = pd.DataFrame(scaler_diabetes.transform(X_diabetes_test), columns=X_diabetes_test.columns)

print("X_diabetes_train_scaled head:")
print(X_diabetes_train_scaled.head())
print("\nX_diabetes_test_scaled head:")
print(X_diabetes_test_scaled.head())
```

## Train Regression Models for Diabetes Dataset

### Subtask:
Train Linear, Lasso (with `alpha=1`), and Ridge (with `alpha=1`) regression models on the scaled diabetes training data.

**Reasoning**:
The subtask requires training Linear, Lasso, and Ridge regression models. This involves importing the respective classes, instantiating each model, setting the `alpha` parameter to 1 for Lasso and Ridge, and fitting them to the scaled training data (`X_diabetes_train_scaled`, `y_diabetes_train`).



```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge

# Instantiate and train Linear Regression model
linear_model = LinearRegression()
linear_model.fit(X_diabetes_train_scaled, y_diabetes_train)
print("Linear Regression model trained successfully.")

# Instantiate and train Lasso Regression model (alpha=1)
lasso_model = Lasso(alpha=1)
lasso_model.fit(X_diabetes_train_scaled, y_diabetes_train)
print("Lasso Regression model trained successfully.")

# Instantiate and train Ridge Regression model (alpha=1)
ridge_model = Ridge(alpha=1)
ridge_model.fit(X_diabetes_train_scaled, y_diabetes_train)
print("Ridge Regression model trained successfully.")
```

## Evaluate Regression Models for Diabetes Dataset

### Subtask:
Evaluate the performance of the trained Linear, Lasso, and Ridge regression models on the scaled diabetes test data using Mean Squared Error and print their coefficients.

**Reasoning**:
The subtask requires evaluating the performance of the trained regression models. This involves making predictions on the scaled test data, calculating the Mean Squared Error for each model, and printing their coefficients.



```python
from sklearn.metrics import mean_squared_error

# Evaluate Linear Regression model
y_pred_linear = linear_model.predict(X_diabetes_test_scaled)
mse_linear = mean_squared_error(y_diabetes_test, y_pred_linear)
print(f"Linear Regression - Mean Squared Error: {mse_linear:.4f}")
print("Linear Regression - Coefficients:", linear_model.coef_)

# Evaluate Lasso Regression model
y_pred_lasso = lasso_model.predict(X_diabetes_test_scaled)
mse_lasso = mean_squared_error(y_diabetes_test, y_pred_lasso)
print(f"\nLasso Regression (alpha=1) - Mean Squared Error: {mse_lasso:.4f}")
print("Lasso Regression (alpha=1) - Coefficients:", lasso_model.coef_)

# Evaluate Ridge Regression model
y_pred_ridge = ridge_model.predict(X_diabetes_test_scaled)
mse_ridge = mean_squared_error(y_diabetes_test, y_pred_ridge)
print(f"\nRidge Regression (alpha=1) - Mean Squared Error: {mse_ridge:.4f}")
print("Ridge Regression (alpha=1) - Coefficients:", ridge_model.coef_)
```
