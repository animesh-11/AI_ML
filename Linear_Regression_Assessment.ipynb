# Task
Load the "College.xlsx" dataset, remove the column containing university names, calculate and round descriptive statistics for numeric columns to 2 decimal places, one-hot encode the 'Private' variable with `drop_first=True`, define 'Grad.Rate' as the target variable, split the data into 70% training and 30% testing sets using `random_state=42`, fit a regression model, output the model coefficients rounded to 2 decimal places, obtain predictions for both training and testing sets, and calculate and round R², Mean Squared Error, and Mean Absolute Error for both training and testing sets to 2 decimal places.

## data_loading

### Subtask:
Load the `College.xlsx` dataset using pandas and remove the column containing university names.


**Reasoning**:
I need to import the pandas library, load the 'College.xlsx' dataset, and display its head to identify the column containing university names as per the instructions.



```python
import pandas as pd

df = pd.read_excel('/content/College.xlsx')
print("Original DataFrame head:")
print(df.head())
```

**Reasoning**:
The previous step identified 'Unnamed: 0' as the column containing university names. Now, I will remove this column from the DataFrame and display the head to confirm the change.



```python
df = df.drop('Unnamed: 0', axis=1)
print("DataFrame head after removing 'Unnamed: 0' column:")
print(df.head())
```

## descriptive_statistics

### Subtask:
Calculate and round descriptive statistics for numeric columns to 2 decimal places.


**Reasoning**:
I need to calculate descriptive statistics for all numeric columns in the DataFrame `df` and then round the results to two decimal places, as specified in the subtask.



```python
descriptive_stats = df.describe().round(2)
print("Descriptive Statistics (rounded to 2 decimal places):")
print(descriptive_stats)
```

## feature_engineering

### Subtask:
One-hot encode the 'Private' variable using `pandas.get_dummies()` with `drop_first=True`.


**Reasoning**:
I need to one-hot encode the 'Private' column using `pd.get_dummies()` with `drop_first=True`, then concatenate the results back to the original DataFrame and drop the original 'Private' column, and finally display the head of the updated DataFrame.



```python
private_dummies = pd.get_dummies(df['Private'], drop_first=True, dtype=int)
df = pd.concat([df, private_dummies], axis=1)
df = df.drop('Private', axis=1)
print("DataFrame head after one-hot encoding 'Private' and dropping original column:")
print(df.head())
```

## data_splitting

### Subtask:
Define 'Grad.Rate' as the target variable and split the data into 70% training and 30% testing sets using `sklearn.model_selection.train_test_split` with `random_state=42`.


**Reasoning**:
I need to import the `train_test_split` function, define the target variable and features, and then split the data into training and testing sets as specified in the instructions. Printing the shapes of the resulting datasets will confirm the split was successful.



```python
from sklearn.model_selection import train_test_split

y = df['Grad.Rate']
X = df.drop('Grad.Rate', axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("Shapes of the datasets after splitting:")
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

## model_training

### Subtask:
Fit a regression model (e.g., Linear Regression) on the training data.


**Reasoning**:
I need to import the `LinearRegression` model from `sklearn.linear_model`, instantiate it, and then fit it to the training data `X_train` and `y_train`.



```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)

print("Linear Regression model fitted successfully.")
```

## coefficient_output

### Subtask:
Output the model coefficients, rounded to 2 decimal places.


**Reasoning**:
I need to extract the coefficients from the fitted model, associate them with their respective feature names, round them to two decimal places, and then print them.



```python
coefficients = pd.Series(model.coef_, index=X_train.columns).round(2)
print("Model Coefficients (rounded to 2 decimal places):")
print(coefficients)
```

## model_prediction

### Subtask:
Obtain predictions for both training and testing sets using the fitted model.

## model_prediction

### Subtask:
Obtain predictions for both training and testing sets using the fitted model.

**Reasoning**:
Now that the model coefficients have been outputted, the next logical step is to use the fitted model to generate predictions on both the training and testing datasets.



```python
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

print("Predictions generated for training and testing sets.")
```

## model_evaluation

### Subtask:
Calculate and round R², Mean Squared Error, and Mean Absolute Error for both training and testing sets to 2 decimal places.

## model_evaluation

### Subtask:
Calculate and round R², Mean Squared Error, and Mean Absolute Error for both training and testing sets to 2 decimal places.

**Reasoning**:
Now that the predictions are available, I need to calculate the R², Mean Squared Error (MSE), and Mean Absolute Error (MAE) for both the training and testing sets, rounding each metric to two decimal places, as specified in the subtask.



```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# Calculate metrics for training set
r2_train = round(r2_score(y_train, y_train_pred), 2)
mse_train = round(mean_squared_error(y_train, y_train_pred), 2)
mae_train = round(mean_absolute_error(y_train, y_train_pred), 2)

# Calculate metrics for testing set
r2_test = round(r2_score(y_test, y_test_pred), 2)
mse_test = round(mean_squared_error(y_test, y_test_pred), 2)
mae_test = round(mean_absolute_error(y_test, y_test_pred), 2)

print("Model Evaluation Metrics:")
print(f"  Training R²: {r2_train}")
print(f"  Training MSE: {mse_train}")
print(f"  Training MAE: {mae_train}")
print(f"  Testing R²: {r2_test}")
print(f"  Testing MSE: {mse_test}")
print(f"  Testing MAE: {mae_test}")

# Get and print the intercept term
intercept = round(model.intercept_, 2)
print(f"\nIntercept Term (rounded to 2 decimal places): {intercept}")
```

**Reasoning**:
The previous code failed because the `.round(2)` method was incorrectly applied directly to the output of `r2_score`, `mean_squared_error`, and `mean_absolute_error`, which are float values. I need to use the built-in `round()` function on the results of these metrics.



```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# Calculate metrics for training set
r2_train = round(r2_score(y_train, y_train_pred), 2)
mse_train = round(mean_squared_error(y_train, y_train_pred), 2)
mae_train = round(mean_absolute_error(y_train, y_train_pred), 2)

# Calculate metrics for testing set
r2_test = round(r2_score(y_test, y_test_pred), 2)
mse_test = round(mean_squared_error(y_test, y_test_pred), 2)
mae_test = round(mean_absolute_error(y_test, y_test_pred), 2)

print("Model Evaluation Metrics:")
print(f"  Training R²: {r2_train}")
print(f"  Training MSE: {mse_train}")
print(f"  Training MAE: {mae_train}")
print(f"  Testing R²: {r2_test}")
print(f"  Testing MSE: {mse_test}")
print(f"  Testing MAE: {mae_test}")
```
