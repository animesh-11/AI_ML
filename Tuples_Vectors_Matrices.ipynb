A regional weather-monitoring center records temperatures every three hours to track short-term climate shifts. Forecasters must classify each 3-hour interval as warming, cooling, fluctuating, or extreme weather when a sudden swing (more than 5°C) occurs. This helps meteorologists issue alerts for rapid shifts while also tracking general temperature patterns.


Write a Python function identify_trends(temperatures) that takes a list of tuples, where each tuple contains two numbers representing the minimum and maximum temperature recorded during a 3-hour block. The function should return a list of labels for each 3-hour block after the first.


The labels are determined as follows:


If the absolute change in either minimum temperature or maximum temperature from the previous block is greater than 5°C, label it as "Extreme Weather"
If both minimum temperature and maximum temperature increased (and neither changed more than 5°C), label it as "Warming"
If both minimum temperature and maximum temperature** decreased (and neither changed more than 5°C), label it as "Cooling"
In all other cases, label it as "Fluctuating"
If the input is not a valid list of tuples (each with exactly two numeric values), or contains fewer than two entries, return the string 'Invalid temperature data'.


Input format


temperatures (list of tuples): each tuple contains two numeric values (float or int)

Output format


A list of str describing the trend for each 3-hour block after the first
Or the str 'Invalid temperature data' if the input is invalid

Constraints

N/A

Example case 1


Input


[(10.0, 25.5), (12.2, 28.1), (18.4, 34.0), (13.0, 31.6)]


Output



['Warming', 'Extreme Weather', 'Extreme Weather']


Example case 2


Input


[(5, 15), (4, 14), (3, 16), (9.2, 22.3)]


Output


['Cooling', 'Fluctuating', 'Extreme Weather']

```python
# Import literal_eval to safely evaluate string input as a Python literal
from ast import literal_eval

def identify_trends(temperatures_str):
    try:
        temperatures = literal_eval(temperatures_str)
    except (ValueError, SyntaxError):
        return 'Invalid temperature data'

    # Check for invalid input
    if not isinstance(temperatures, list) or len(temperatures) < 2:
        return 'Invalid temperature data'
    for block in temperatures:
        if not isinstance(block, tuple) or len(block) != 2:
            return 'Invalid temperature data'
        for temp in block:
            if not isinstance(temp, (int, float)):
                return 'Invalid temperature data'

    trends = []
    for i in range(1, len(temperatures)):
        prev_min, prev_max = temperatures[i-1]
        curr_min, curr_max = temperatures[i]

        min_diff = curr_min - prev_min
        max_diff = curr_max - prev_max

        if abs(min_diff) > 5 or abs(max_diff) > 5:
            trends.append("Extreme Weather")
        elif min_diff > 0 and max_diff > 0:
            trends.append("Warming")
        elif min_diff < 0 and max_diff < 0:
            trends.append("Cooling")
        else:
            trends.append("Fluctuating")

    return trends

# Example Usage (you can replace this with your actual input)
print(identify_trends("[(10.0, 25.5), (12.2, 28.1), (18.4, 34.0), (13.0, 31.6)]"))
print(identify_trends("[(5, 15), (4, 14), (3, 16), (9.2, 22.3)]"))
print(identify_trends("[]"))
print(identify_trends("[(10, 20)]"))
print(identify_trends("[(10, 20, 30)]"))
print(identify_trends('[("a", 20), (10, 30)]'))
print(identify_trends("invalid input"))
```

A retail store keeps track of customer purchases in a list of tuples. Understanding customer purchase behavior is key for optimising sales strategies and enhancing the customer experience. By analysing this data, the store can provide targeted promotions, manage inventory efficiently, and improve customer retention.


Your task is to write a function get_customer_purchase_count(purchases) that:


Takes a list of tuples as input, where each tuple contains the customer ID and the product name they purchased
Returns a dictionary of dictionaries:
The outer dictionary’s keys are customer IDs
The values in the outer dictionary are inner dictionaries, where:
The inner dictionary’s keys are product names
The values represent how many times the corresponding customer purchased each product

Input Format


A list of tuples, where each tuple represents a transaction:
The first element is the customer ID (str)
The second element is the product name (str)

Output Format


A dictionary of dictionaries, where the outer dictionary has customer IDs as keys and the inner dictionary contains:
Product names as keys
The number of times each product was purchased by that customer as values

Constraints


N/A

Example Case 1


Input


[('C001', 'Laptop'), ('C002', 'Phone'), ('C001', 'Laptop'), ('C001', 'Headphones'), ('C002', 'Phone')]


Output


{'C001': {'Laptop': 2, 'Headphones': 1}, 'C002': {'Phone': 2}}


Example Case 2


Input


[('C001', 'Socks'), ('C001', 'Shoes'), ('C002', 'Socks'), ('C003', 'Shoes'), ('C003', 'Socks')]


Output


{'C001': {'Socks': 1, 'Shoes': 1}, 'C002': {'Socks': 1}, 'C003': {'Shoes': 1, 'Socks': 1}}

```python
from ast import literal_eval

def get_customer_purchase_count(purchases_str):
    try:
        purchases = literal_eval(purchases_str)
    except (ValueError, SyntaxError):
        return "Invalid input data" # Or handle error as appropriate

    # Check if the input is a list of tuples
    if not isinstance(purchases, list) or not all(isinstance(p, tuple) and len(p) == 2 for p in purchases):
         return "Invalid input data" # Or handle error as appropriate


    # Create a regular dictionary to store purchase counts for each customer and product
    customer_purchases = {}

    # Iterate through the list of purchases
    for customer_id, product_name in purchases:
        # If the customer_id is not in the dictionary, add it with an empty dictionary as the value
        if customer_id not in customer_purchases:
            customer_purchases[customer_id] = {}

        # If the product_name is not in the inner dictionary for the customer, add it with a count of 0
        if product_name not in customer_purchases[customer_id]:
            customer_purchases[customer_id][product_name] = 0

        # Increment the count for the specific customer and product
        customer_purchases[customer_id][product_name] += 1

    return customer_purchases

# Example Usage
purchases_str1 = "[('C001', 'Laptop'), ('C002', 'Phone'), ('C001', 'Laptop'), ('C001', 'Headphones'), ('C002', 'Phone')]"
print(get_customer_purchase_count(purchases_str1))

purchases_str2 = "[('C001', 'Socks'), ('C001', 'Shoes'), ('C002', 'Socks'), ('C003', 'Shoes'), ('C003', 'Socks')]"
print(get_customer_purchase_count(purchases_str2))

# Example with invalid input
invalid_purchases_str = "[('C001', 'Laptop'), ('C002')]"
print(get_customer_purchase_count(invalid_purchases_str))
```

A streaming platform wants to recommend movies by comparing a user’s typical genre-specific ratings with how other viewers have rated each genre of every movie


Each user has a genre rating profile — a list of numbers where each number represents how much the user typically rates movies from a specific genre. For example, if the genres are ['Action', 'Comedy', 'Suspense'], a user vector [3, 2, 5] means the user usually gives Action movies a 3, Comedy a 2, and Suspense a 5.


Each movie is represented by a genre vector — a list of numbers indicating how highly other viewers rated each genre for that movie. For example, a movie vector [3, 1, 5] means this movie is considered strong in Action and Suspense, but weak in Comedy.


To decide which movie a user will most likely enjoy, the platform computes a recommendation score using the dot product of the user’s vector and each movie’s vector. The higher the score, the stronger the match.


Write a Python function recommend_movie(user_vector, movie_vectors) that:


Accepts a user vector and a list of movie vectors
Computes the dot product for each movie
Returns the 1-based index of the movie with the highest score
If multiple movies tie for the highest score, return the one with the lowest index
If any movie vector is not the same length as the user vector, return 'Invalid movie vector(s)'

Note: The output should use 1-based indexing. This means that the first movie in the list is referred to as Movie 1, the second as Movie 2, and so on — even though Python lists are 0-based internally.

For example, if the movie at index 1 in the list (i.e. the second movie) has the highest score, the output should be: 'Movie 2 is highly likely to be enjoyed by the user.'

Input format


user_vector (list): Represents the user’s typical rating (float or int) for each genre. The i-th element corresponds to the user’s rating for the i-th genre.
movie_vectors (list of lists): Each inner list represents one movie. The i-th value in a movie vector is the average rating (float or int) from other users that the movie received for the i-th genre.

Output format


A str 'Movie X is highly likely to be enjoyed by the user', where X is the 1-based index of the recommended movie
Or a str 'Invalid movie vector(s)' if any movie vector is not the same length as the user vector

Constraints


All ratings are non-negative numbers
The input list contains at least one movie

Example Case 1


Inputs

[3, 2, 5]

[[3, 1, 5], [2, 3, 1], [0, 0, 4]]



Output


Movie 1 is highly likely to be enjoyed by the user


Example Case 2


Inputs

 [1, 2, 3]

 [[2, 2], [1, 2, 3], [0, 0, 9]]



Output

Invalid movie vector(s)

```python
def recommend_movie(user_vector, movie_vectors):
    user_vector_length = len(user_vector)
    highest_score = -1
    recommended_movie_index = -1

    for i, movie_vector in enumerate(movie_vectors):
        # Check if the movie vector has the same length as the user vector
        if len(movie_vector) != user_vector_length:
            return 'Invalid movie vector(s)'

        # Compute the dot product
        dot_product = sum(u * m for u, m in zip(user_vector, movie_vector))

        # Check if the current movie has a higher score or is the first highest score
        if dot_product > highest_score:
            highest_score = dot_product
            recommended_movie_index = i

    # Return the result using 1-based indexing
    return f'Movie {recommended_movie_index + 1} is highly likely to be enjoyed by the user'

# Example Case 1
user_vector_1 = [3, 2, 5]
movie_vectors_1 = [[3, 1, 5], [2, 3, 1], [0, 0, 4]]
print(recommend_movie(user_vector_1, movie_vectors_1))

# Example Case 2
user_vector_2 = [1, 2, 3]
movie_vectors_2 = [[2, 2], [1, 2, 3], [0, 0, 9]]
print(recommend_movie(user_vector_2, movie_vectors_2))
```

In image processing, a common task is to convert a coloured image into greyscale. Each pixel in an image is represented by three values (red, green, and blue channels), where each value is in the range [0,255]. To convert an image to greyscale, a typical approach is to calculate the average of the red, green, and blue values of each pixel.


For instance, for each pixel, the greyscale value is calculated as:


greyscale = (red + green + blue)/3



Write a Python function convert_to_greyscale(image) that takes an image matrix as input, where the image is represented by a list of lists, and each element in the matrix is a list containing three integers representing the red, green, and blue values of a pixel. The function should return a new matrix, where each element is the greyscale value of the pixel calculated as the average of the red, green, and blue values.



Input format


image: A list of lists, where each inner list contains three values (int) representing the red, green, and blue values of a pixel of the image; there is no restriction on the length of the outer list

Output format


If any red, green, and blue values are not in the range [0,255], return a string 'Invalid RGB values present'
Otherwise, return a list where each element (int) represents the greyscale value (average of the red, green, and blue values, rounded down) of the corresponding pixel in the original image

Constraints


The elements within the input list must be lists three elements long

Example case 1


Input


[[255, 0, 0], [0, 255, 0], [0, 0, 255]]


Output

[85, 85, 85]


Example case 2


Input


[[-1, 0, 255], [0, 0, 255]]


Output


Invalid RGB values present

```python
def convert_to_greyscale(image):
    greyscale_image = []
    for pixel in image:
        if len(pixel) != 3:
            return 'Invalid RGB values present'  # Constraint check
        r, g, b = pixel
        if not all(0 <= val <= 255 for val in [r, g, b]):
            return 'Invalid RGB values present'
        greyscale_value = (r + g + b) // 3  # Integer division for rounding down
        greyscale_image.append(greyscale_value)
    return greyscale_image

# Example Case 1
image1 = [[255, 0, 0], [0, 255, 0], [0, 0, 255]]
print(convert_to_greyscale(image1))

# Example Case 2
image2 = [[-1, 0, 255], [0, 0, 255]]
print(convert_to_greyscale(image2))

# Example with invalid input format
image3 = [[255, 0], [0, 255, 0]]
print(convert_to_greyscale(image3))
```

In the logistics and transportation industry, forces acting on cargo shipments are critical to understanding how shipments will behave under various conditions. For example, wind forces, road resistance, and driver inputs create vectors that affect the shipment's motion. The resultant force is the vector sum of all individual forces acting on the cargo, and scaling the resultant force can help optimise operations for different cargo sizes and transport conditions.


Write a Python function resultant_and_scaled_force(v1, v2, scalar) that takes:


Two force vectors, each represented by a tuple of three integers, where the first integer is the x-component, the second is the y-component, and the third is the z-component
A scalar integer representing the scaling factor (for example, scaling due to changes in cargo weight, operational adjustments, or environmental factors)

The function should return the scaled resultant force vector obtained by:


Adding the two vectors to get the resultant vector
Multiplying the resultant vector by the scalar factor to adjust for the scaling needs

The output should be a tuple representing the x, y, and z components of the scaled resultant vector


Input format

v1: a tuple containing three numbers (int) representing the x, y, and z components of the first vector
v2 a tuple containing three numbers (int) representing the x, y, and z components of the second vector
scalar: a scalar (int)

Output format

A tuple containing the x, y, and z components (int) of the scaled resultant force vector

Constraints


N/A

Example case 1

Inputs



(3, 4, 5)

(1, 2, 3)

2


Output


(8, 12, 16)



Example case 2


Inputs



(-1, -2, -3)

(5, 6, 7)

3


Output

(12, 12, 12)

```python
def resultant_and_scaled_force(v1, v2, scalar):
    # Calculate the resultant vector
    resultant_vector = (v1[0] + v2[0], v1[1] + v2[1], v1[2] + v2[2])

    # Scale the resultant vector
    scaled_resultant_vector = (resultant_vector[0] * scalar,
                               resultant_vector[1] * scalar,
                               resultant_vector[2] * scalar)

    return scaled_resultant_vector

# Example Case 1
v1_1 = (3, 4, 5)
v2_1 = (1, 2, 3)
scalar_1 = 2
print(resultant_and_scaled_force(v1_1, v2_1, scalar_1))

# Example Case 2
v1_2 = (-1, -2, -3)
v2_2 = (5, 6, 7)
scalar_2 = 3
print(resultant_and_scaled_force(v1_2, v2_2, scalar_2))
```

To analyse route efficiency, the system must sort these vectors based on how far each vehicle has travelled — that is, by the magnitude of the vector.


Write a Python function sort_by_magnitude(vectors) that:


Accepts a list of 2-D vectors (tuples of two int values)
Computes each vector’s magnitude using the formula above
Returns a new list of vectors sorted in descending order of magnitude
If two vectors have the same magnitude, keep them in their original input order

Input format

vectors (list of tuples of int): each element in the list is a tuple (x, y) representing a 2-D vector

Output format

A list of tuples of int, sorted in descending order by magnitude
Vectors with equal magnitude remain in their original order

Constraints

The input list contains at least one tuple
Each tuple contains exactly two numbers (int)
Use the built-in sorted() function with the key parameter to specify that sorting should be based on each vector’s magnitude. A lambda function is commonly used here to define that rule inline

Example case 1

Input


[(3, 4), (1, 1), (0, 2)]


Output

[(3, 4), (0, 2), (1, 1)]

```python
def sort_by_magnitude(vectors):
  """
  Sorts a list of 2D vectors by magnitude in descending order.

  Args:
    vectors: A list of tuples, where each tuple is a 2D vector (x, y).

  Returns:
    A new list of vectors sorted by magnitude in descending order.
    Vectors with equal magnitude remain in their original order.
  """
  # Use a lambda function as the key for sorting by magnitude
  # The magnitude is calculated as the square root of (x^2 + y^2)
  # We use -((v[0]**2 + v[1]**2)**0.5) to sort in descending order
  sorted_vectors = sorted(vectors, key=lambda v: -(v[0]**2 + v[1]**2)**0.5)
  return sorted_vectors

# Example Case 1
vectors1 = [(3, 4), (1, 1), (0, 2)]
print(sort_by_magnitude(vectors1))
```

Your task is to combine the readings from different sensors to create a final 3D vector (light intensity, temperature, and humidity) by filling in the missing data. If more than one sensor provides the same measurement for a particular factor, the system must calculate the average value of that factor.

You need to write a function combine_sensor_data(sensors_data) that:


Takes a list of dictionaries where each dictionary represents a sensor reading containing at least one of the three factors: 'light intensity', 'temperature', and 'humidity'
The function should combine the readings and return the final vector (light, intensity, temperature, humidity)
If a sensor reading does not contain one of the factors, use the average of all the available readings to fill in the missing factor
Ensure that if more than one reading provides the same factor (e.g., multiple readings for light intensity), the average of those readings is used for that factor

Input Format


A list of dictionaries, where each dictionary contains at least one of the following three possible keys:
'light intensity' (int)
'temperature' (int)
'humidity' (int)

Output Format


A tuple with three values: (light intensity, temperature, humidity), where:
The light intensity (float) is the average (rounding to 2 decimal places) of all available readings for this factor
The temperature (float) is the average (rounding to 2 decimal places) of all available readings for this factor
The humidity (float) is the average (rounding to 2 decimal places) of all available readings for this factor

Constraints


Sensor readings for light intensity, temperature, and humidity are always non-negative

Example Case 1

Input

[{'light intensity': 10, 'temperature': 300,'humidity': 100}, {'light intensity': 20, 'humidity': 75}, {'temperature': 280, 'humidity': 80}]

Input


[{'light intensity': 100, 'temperature': 200}, {'humidity': 50, 'temperature': 220}, {'light intensity': 150, 'humidity': 40}]

```python
def combine_sensor_data(sensors_data):
    light_readings = []
    temp_readings = []
    humidity_readings = []

    for sensor in sensors_data:
        if 'light intensity' in sensor:
            light_readings.append(sensor['light intensity'])
        if 'temperature' in sensor:
            temp_readings.append(sensor['temperature'])
        if 'humidity' in sensor:
            humidity_readings.append(sensor['humidity'])

    # Calculate averages, handle cases with no readings
    avg_light = sum(light_readings) / len(light_readings) if light_readings else 0.0
    avg_temp = sum(temp_readings) / len(temp_readings) if temp_readings else 0.0
    avg_humidity = sum(humidity_readings) / len(humidity_readings) if humidity_readings else 0.0

    # Round to 2 decimal places
    avg_light = round(avg_light, 2)
    avg_temp = round(avg_temp, 2)
    avg_humidity = round(avg_humidity, 2)

    return (avg_light, avg_temp, avg_humidity)

# Example Case 1
sensors_data1 = [{'light intensity': 10, 'temperature': 300,'humidity': 100}, {'light intensity': 20, 'humidity': 75}, {'temperature': 280, 'humidity': 80}]
print(combine_sensor_data(sensors_data1))

# Example Case 2
sensors_data2 = [{'light intensity': 100, 'temperature': 200}, {'humidity': 50, 'temperature': 220}, {'light intensity': 150, 'humidity': 40}]
print(combine_sensor_data(sensors_data2))

# Example with missing values and only one reading for a factor
sensors_data3 = [{'light intensity': 50}, {'temperature': 250, 'humidity': 60}]
print(combine_sensor_data(sensors_data3))

# Example with an empty list
sensors_data4 = []
print(combine_sensor_data(sensors_data4))
```
