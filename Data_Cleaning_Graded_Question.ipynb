```python
def clean_phone_numbers(contacts_dict):
    cleaned_dict = {}
    for name, phone_number in contacts_dict.items():
        # Remove all non-digit characters
        cleaned_number = ''.join(filter(str.isdigit, phone_number))

        # Check if the cleaned number has exactly 10 digits
        if len(cleaned_number) == 10:
            cleaned_dict[name] = cleaned_number
        else:
            # If not 10 digits, keep the original phone number
            cleaned_dict[name] = phone_number
    return cleaned_dict


# Example Testcase 1
input_1 = {'Aisha': '(123) 456-7890', 'Bipin': '+123 654 3210', 'Nina': ' 000  '}
expected_output_1 = {'Aisha': '1234567890', 'Bipin': '1236543210', 'Nina': ' 000  '}
cleaned_output_1 = clean_phone_numbers(input_1)

print(f"Input 1: {input_1}")
print(f"Expected Output 1: {expected_output_1}")
print(f"Cleaned Output 1: {cleaned_output_1}")
print(f"Test 1 Passed: {cleaned_output_1 == expected_output_1}\n")

# Example Testcase 2
input_2 = {'Mansi': '123__ 9834 - 435', 'Tushar': '176 _ - 7576', 'Parul': '+1 (800) 555-1234'}
expected_output_2 = {'Mansi': '1239834435', 'Tushar': '176 _ - 7576', 'Parul': '+1 (800) 555-1234'}
cleaned_output_2 = clean_phone_numbers(input_2)

print(f"Input 2: {input_2}")
print(f"Expected Output 2: {expected_output_2}")
print(f"Cleaned Output 2: {cleaned_output_2}")
print(f"Test 2 Passed: {cleaned_output_2 == expected_output_2}")
```

## Define outlier detection and removal function

### Subtask:
Create a Python function that uses the IQR method to identify outliers in specified columns and removes the corresponding rows from the DataFrame.


**Reasoning**:
I need to define a Python function `remove_outliers_iqr` that takes a DataFrame and a list of columns. This function will calculate the IQR for each specified column and remove rows that contain outliers based on the 1.5 * IQR rule.



```python
def remove_outliers_iqr(df, columns):
    cleaned_df = df.copy()
    for col in columns:
        if col in cleaned_df.columns:
            Q1 = cleaned_df[col].quantile(0.25)
            Q3 = cleaned_df[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            cleaned_df = cleaned_df[(cleaned_df[col] >= lower_bound) & (cleaned_df[col] <= upper_bound)]
    return cleaned_df

print("Function `remove_outliers_iqr` defined successfully.")
```

## Apply outlier removal

### Subtask:
Apply the defined function to the 'temperature', 'humidity', and 'air_quality' columns of the DataFrame.


**Reasoning**:
I will apply the `remove_outliers_iqr` function to the specified columns of the DataFrame and store the result in a new DataFrame called `df_cleaned`.



```python
columns_to_clean = ['temperature', 'humidity', 'air_quality']
df_cleaned = remove_outliers_iqr(df, columns_to_clean)

print(f"Original DataFrame shape: {df.shape}")
print(f"Cleaned DataFrame shape: {df_cleaned.shape}")
df_cleaned.head()
```
