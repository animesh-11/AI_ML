Given a scalar function







Write a program that computes the gradient vector
Input format

Two real numbers in a tuple: (x, y)



Output format

List containing the gradient at (x,y), rounded to 2 decimal places





Sample input

(1, 1)



Sample output

[4.0, 4.0]

```python
import math

def compute_gradient(point):
  """
  Computes the gradient of the function f(x, y) = x^2 + y^2 + 2xy at a given point (x, y)
  using only the math library.

  Args:
    point: A tuple containing the x and y coordinates (x, y).

  Returns:
    A list containing the gradient vector at the given point, rounded to 2 decimal places.
  """
  x, y = point

  # Partial derivative with respect to x: d/dx (x^2 + y^2 + 2xy) = 2x + 2y
  df_dx = 2 * x + 2 * y

  # Partial derivative with respect to y: d/dy (x^2 + y^2 + 2xy) = 2y + 2x
  df_dy = 2 * y + 2 * x

  return [round(df_dx, 2), round(df_dy, 2)]

# Sample input
sample_input = (1, 1)
sample_output = compute_gradient(sample_input)
print(f"The gradient at {sample_input} is: {sample_output}")
```

You are given a list of outcomes and their respective probabilities as two lists. Write a Python function to compute the expected value of the random variable.
If the sum of probabilities is not 1, return "Invalid input".





Input Format:



A list containing two lists:



[ [x1, x2, ..., xn], [p1, p2, ..., pn] ]



The outcomes and probabilities lists have been extracted already in the starter code.

outcomes = [x1, x2, ..., xn]

probabilities = [p1, p2, ..., pn]



Output Format:

Expected Value: A float rounded to 2 decimal places or "Invalid input"





Sample Input:

[[1, 2, 3], [0.2, 0.5, 0.3]]



Sample Output:

2.1

```python
def compute_expected_value(data):
  """
  Computes the expected value of a random variable given outcomes and probabilities.

  Args:
    data: A list containing two lists: [[outcomes], [probabilities]].

  Returns:
    The expected value rounded to 2 decimal places, or "Invalid input"
    if the sum of probabilities is not 1.
  """
  outcomes = data[0]
  probabilities = data[1]

  if round(sum(probabilities), 2) != 1.0:
    return "Invalid input"

  expected_value = sum(x * p for x, p in zip(outcomes, probabilities))
  return round(expected_value, 2)

# Sample Input
sample_input = [[1, 2, 3], [0.2, 0.5, 0.3]]
sample_output = compute_expected_value(sample_input)
print(f"The expected value for {sample_input} is: {sample_output}")
```

You are given a normally distributed variable with mean μ and standard deviation σ, and a value x.

Write a Python function to calculate the corresponding z-score
Round the result to 2 decimal places. If σ ≤ 0, return "Invalid input".



Input Format:

A tuple: (x, mu (μ), sigma (σ) )



Output Format:

A float rounded to 2 decimal places or "Invalid input"





Sample Input:

(75, 70, 5)



Sample Output:

1.0

```python
def calculate_z_score(data):
  """
  Calculates the z-score for a given value x, mean (μ), and standard deviation (σ).

  Args:
    data: A tuple containing (x, μ, σ).

  Returns:
    The z-score rounded to 2 decimal places, or "Invalid input" if σ is not positive
    or if inputs are not numerical.
  """
  x, mu, sigma = data

  # Check if inputs are numerical
  if not all(isinstance(i, (int, float)) for i in [x, mu, sigma]):
    return "Invalid input"

  if sigma <= 0:
    return "Invalid input"

  z_score = (x - mu) / sigma
  return round(z_score, 2)

# Sample Input
sample_input = (75, 70, 5)
sample_output = calculate_z_score(sample_input)
print(f"The z-score for {sample_input} is: {sample_output}")

# Sample edge case input (non-numerical)
sample_input_edge = ("abc", 70, 5)
sample_output_edge = calculate_z_score(sample_input_edge)
print(f"The z-score for {sample_input_edge} is: {sample_output_edge}")

# Sample edge case input (sigma <= 0)
sample_input_edge_sigma = (75, 70, 0)
sample_output_edge_sigma = calculate_z_score(sample_input_edge_sigma)
print(f"The z-score for {sample_input_edge_sigma} is: {sample_output_edge_sigma}")
```

Given two vectors u and v in Rn space, write a Python function that:

Computes the magnitude of each vector
Computes the dot product of the two vectors


If the vectors are not of the same length, return "Invalid input".





Input Format:

A list containing two lists of floats: [list u, list v]

Note that the length of the strings is not fixed



Output Format:

A tuple with three floats:

(||u||, ||v||, u ⋅ v) magnitudes of both vectors and their dot product, rounded to 2 decimal places.




Or "Invalid input" if dot product is not possible between the two vectors





Sample Input:

[[1, 2, 3], [4, 5, 6]]



Sample Output:

(3.74, 8.77, 32.0)

```python
import math

def vector_operations(data):
  """
  Computes the magnitude of two vectors and their dot product.

  Args:
    data: A list containing two lists of floats: [list u, list v].

  Returns:
    A tuple with three floats: (||u||, ||v||, u ⋅ v) rounded to 2 decimal places,
    or "Invalid input" if the vectors are not of the same length or contain non-numerical values.
  """
  if not isinstance(data, list) or len(data) != 2:
      return "Invalid input"

  u, v = data

  if not isinstance(u, list) or not isinstance(v, list):
      return "Invalid input"

  if len(u) != len(v):
    return "Invalid input"

  # Check for non-numerical inputs
  if not all(isinstance(x, (int, float)) for x in u) or not all(isinstance(y, (int, float)) for y in v):
      return "Invalid input"

  # Compute magnitude of u
  magnitude_u = math.sqrt(sum(x**2 for x in u))

  # Compute magnitude of v
  magnitude_v = math.sqrt(sum(y**2 for y in v))

  # Compute dot product
  dot_product = sum(x * y for x, y in zip(u, v))

  return (round(magnitude_u, 2), round(magnitude_v, 2), round(dot_product, 2))

# Sample Input
sample_input = [[1, 2, 3], [4, 5, 6]]
sample_output = vector_operations(sample_input)
print(f"For vectors {sample_input[0]} and {sample_input[1]}, the magnitudes and dot product are: {sample_output}")

# Sample edge case input (different lengths)
sample_input_edge_length = [[1, 2], [4, 5, 6]]
sample_output_edge_length = vector_operations(sample_input_edge_length)
print(f"For vectors {sample_input_edge_length[0]} and {sample_input_edge_length[1]}, the result is: {sample_output_edge_length}")

# Sample edge case input (non-numerical)
sample_input_edge_numerical = [[1, "a", 3], [4, 5, 6]]
sample_output_edge_numerical = vector_operations(sample_input_edge_numerical)
print(f"For vectors {sample_input_edge_numerical[0]} and {sample_input_edge_numerical[1]}, the result is: {sample_output_edge_numerical}")

# Sample edge case input (invalid data structure)
sample_input_edge_structure = [1, 2, 3]
sample_output_edge_structure = vector_operations(sample_input_edge_structure)
print(f"For input {sample_input_edge_structure}, the result is: {sample_output_edge_structure}")
```

Given a function







Write a program to compute the partial derivatives



at a given point (x,y).



Return both values rounded to 2 decimal places.





Input format

Two real numbers as a tuple: (x, y)



Output format

Tuple:






Sample input

(1,2)



Sample output

(4.0, 13.0)



Partial derivatives of the function w.r.t. x and y respectively

```python
def compute_partial_derivatives(point):
  """
  Computes the partial derivatives of the function f(x, y) = x^2*y + y^3
  at a given point (x, y).

  Args:
    point: A tuple containing the x and y coordinates (x, y).

  Returns:
    A tuple containing the partial derivatives with respect to x and y,
    rounded to 2 decimal places.
  """
  x, y = point

  # Partial derivative with respect to x: d/dx (x^2*y + y^3) = 2xy
  df_dx = 2 * x * y

  # Partial derivative with respect to y: d/dy (x^2*y + y^3) = x^2 + 3y^2
  df_dy = x**2 + 3 * y**2

  return (round(df_dx, 2), round(df_dy, 2))

# Sample input
sample_input = (1, 2)
sample_output = compute_partial_derivatives(sample_input)
print(f"The partial derivatives at {sample_input} are: {sample_output}")
```
