```python
# Check for Missing Values
from pyspark.sql.functions import col, sum

def check_missing_values(df, df_name):
    print(f"\nMissing values in {df_name} DataFrame:")
    missing_counts = df.select([sum(col(c).isNull().cast("integer")).alias(c) for c in df.columns])
    missing_counts.show(truncate=False)

check_missing_values(collisions_df, "Collisions")
check_missing_values(parties_df, "Parties")
check_missing_values(victims_df, "Victims")
check_missing_values(case_ids_df, "Case_IDs")
```

```python
# Drop Sparse Columns
from pyspark.sql.functions import col, sum

def calculate_sparse_columns(df, df_name, threshold=0.5):

    total_rows = df.count()
    if total_rows == 0:
        return []

    missing_counts_row = df.select([sum(col(c).isNull().cast("integer")).alias(c) for c in df.columns]).collect()[0]
    missing_data = missing_counts_row.asDict()

    sparse_columns = []

    if df_name == "Collisions":
        columns_to_keep_for_df = [c for c in ['latitude', 'longitude'] if c in df.columns]
    else:
        columns_to_keep_for_df = []

    for column, count in missing_data.items():
        if column in columns_to_keep_for_df:
            if count is not None and (count / total_rows) > threshold:
                print("Continue")
            continue
        if count is not None and (count / total_rows) > threshold:
            sparse_columns.append(column)
            print(f"  Column '{column}' has {count} missing values ({count / total_rows:.2%}) and will be dropped.")
    return sparse_columns

def drop_sparse_columns(df, columns_to_drop, df_name):
    if not columns_to_drop:
        print(f"No sparse columns to drop from {df_name} DataFrame.")
        return df

    print(f"\nDropping sparse columns from {df_name} DataFrame: {columns_to_drop}")
    df_cleaned = df.drop(*columns_to_drop)
    df_cleaned.printSchema()
    print(f"Dropped columns from {df_name}: {', '.join(columns_to_drop)}")
    return df_cleaned

# List of DataFrames to process
dataframes_to_process = {
    "Collisions": collisions_df,
    "Parties": parties_df,
    "Victims": victims_df,
    "Case_IDs": case_ids_df
}

cleaned_dataframes = {}

for name, df in dataframes_to_process.items():
    sparse_cols = calculate_sparse_columns(df, name, threshold=0.5)
    cleaned_df = drop_sparse_columns(df, sparse_cols, name)
    cleaned_dataframes[name] = cleaned_df

# Update the original DataFrames with the cleaned ones
collisions_df = cleaned_dataframes["Collisions"]
parties_df = cleaned_dataframes["Parties"]
victims_df = cleaned_dataframes["Victims"]
case_ids_df = cleaned_dataframes["Case_IDs"]
```

```python
#Handle Missing Values

from pyspark.sql.functions import col, sum, median
from pyspark.sql.types import DoubleType


# Function for median imputation
def impute_with_median(df, columns):
    print(f"\nApplying median imputation to {df.alias} DataFrame for columns: {columns}")
    for col_name in columns:
        if col_name in df.columns:
            median_value = df.approxQuantile(col_name, [0.5], 0.01)[0] if df.filter(col(col_name).isNotNull()).count() > 0 else None
            if median_value is not None:
                df = df.fillna({col_name: median_value})
                print(f"  Filled missing values in '{col_name}' with median: {median_value}")
            else:
                print(f"All values are null or column does not exist.")
        else:
            print(f"  Column '{col_name}' not found in DataFrame {df.alias}. Skipping.")
    return df

# Function for mode imputation
def impute_with_mode(df, columns):
    print(f"\nApplying mode imputation to {df.alias} DataFrame for columns: {columns}")
    for col_name in columns:
        if col_name in df.columns:

            mode_df = df.groupBy(col_name).count().orderBy(col("count").desc()).limit(1)
            mode_value = mode_df.collect()[0][col_name] if mode_df.count() > 0 else None

            if mode_value is not None:
                df = df.fillna({col_name: mode_value})
                print(f"  Filled missing values in '{col_name}' with mode: {mode_value}")
            else:
                print(f"All values are null or column does not exist.")
        else:
            print(f"  Column '{col_name}' not found in DataFrame {df.alias}. Skipping.")
    return df


collisions_df = collisions_df.alias("Collisions")
parties_df = parties_df.alias("Parties")
victims_df = victims_df.alias("Victims")
case_ids_df = case_ids_df.alias("Case_IDs")

# Median imputation
median_cols_collisions = ["jurisdiction", "special_condition", "intersection", "state_highway_indicator", "caltrans_district", "state_route", "tow_away", "killed_victims", "injured_victims", "pcf_violation", "postmile"]
#median_cols_collisions = [col_name for col_name in median_cols_collisions if col_name in collisions_df.columns]
collisions_df = impute_with_median(collisions_df, median_cols_collisions)

median_cols_parties = ["party_age", "cellphone_in_use", "vehicle_year"]
#median_cols_parties = [col_name for col_name in median_cols_parties if col_name in parties_df.columns]
parties_df = impute_with_median(parties_df, median_cols_parties)

median_cols_victims = ["victim_age"]
#median_cols_victims = [col_name for col_name in median_cols_victims if col_name in victims_df.columns]
victims_df = impute_with_median(victims_df, median_cols_victims)

# Mode imputation
mode_cols_collisions = ["officer_id", "reporting_district", "population", "chp_beat_class", "beat_number", "primary_road", "secondary_road", "direction", "weather_1", "primary_collision_factor", "pcf_violation_category", "pcf_violation_subsection", "type_of_collision", "motor_vehicle_involved_with", "pedestrian_action", "road_surface", "road_condition_1", "lighting", "control_device", "statewide_vehicle_type_at_fault", "chp_vehicle_type_at_fault"]
#mode_cols_collisions = [col_name for col_name in mode_cols_collisions if col_name in collisions_df.columns]
collisions_df = impute_with_mode(collisions_df, mode_cols_collisions)

mode_cols_parties = ["party_type", "party_sex", "party_sobriety", "direction_of_travel", "party_safety_equipment_1", "party_safety_equipment_2", "financial_responsibility", "cellphone_use_type", "other_associate_factor_1", "movement_preceding_collision", "vehicle_make", "statewide_vehicle_type", "chp_vehicle_type_towing", "chp_vehicle_type_towed", "party_race"]
#mode_cols_parties = [col_name for col_name in mode_cols_parties if col_name in parties_df.columns]
parties_df = impute_with_mode(parties_df, mode_cols_parties)

mode_cols_victims = ["victim_sex", "victim_seating_position", "victim_safety_equipment_1", "victim_safety_equipment_2", "victim_ejected"]
#mode_cols_victims = [col_name for col_name in mode_cols_victims if col_name in victims_df.columns]
victims_df = impute_with_mode(victims_df, mode_cols_victims)


# Drop rows with nulls in collision_date and process_date
date_cols_to_check = ["collision_date", "process_date"]
initial_collision_count = collisions_df.count()
collisions_df = collisions_df.na.drop(subset=date_cols_to_check)
final_collision_count = collisions_df.count()
if initial_collision_count != final_collision_count:
    print(f"Dropped {initial_collision_count - final_collision_count} rows from Collisions DataFrame due to nulls in date columns.")
else:
    print("No rows dropped from Collisions DataFrame based on date columns.")

check_missing_values(collisions_df, "Collisions")
check_missing_values(parties_df, "Parties")
check_missing_values(victims_df, "Victims")
check_missing_values(case_ids_df, "Case_IDs")
```

```python
#Detect Outliers using IQR

from pyspark.sql.functions import col, lit, when
from pyspark.sql.types import IntegerType, DoubleType, LongType

def detect_outliers_iqr(df, column_name, df_name):
    print(f"\n--- Detecting outliers in '{column_name}' of {df_name} DataFrame ---")

    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    quantiles = df.approxQuantile(column_name, [0.25, 0.75], 0.01)
    q1 = quantiles[0]
    q3 = quantiles[1]
    iqr = q3 - q1


    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    # Count outliers
    outliers_count = df.filter((col(column_name) < lower_bound) | (col(column_name) > upper_bound)).count()
    total_count = df.count()

    print(f"  Q1: {q1}, Q3: {q3}, IQR: {iqr}")
    print(f"  Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
    print(f"  Number of outliers in '{column_name}': {outliers_count} ({outliers_count / total_count:.2%} of total rows)")

    return outliers_count

def get_numerical_columns(df):
    """Returns a list of numerical column names from a DataFrame."""
    numerical_cols = []
    for field in df.schema.fields:
        if isinstance(field.dataType, (IntegerType, DoubleType, LongType)):
            numerical_cols.append(field.name)
    return numerical_cols

# Detect outliers for Collisions DataFrame
collisions_numerical_cols = get_numerical_columns(collisions_df)
print(f"\nIdentified numerical columns in Collisions DataFrame: {collisions_numerical_cols}")
for col_name in collisions_numerical_cols:
    detect_outliers_iqr(collisions_df, col_name, "Collisions")

# Detect outliers for Parties DataFrame
parties_numerical_cols = get_numerical_columns(parties_df)
print(f"\nIdentified numerical columns in Parties DataFrame: {parties_numerical_cols}")
for col_name in parties_numerical_cols:
    detect_outliers_iqr(parties_df, col_name, "Parties")

# Detect outliers for Victims DataFrame
victims_numerical_cols = get_numerical_columns(victims_df)
print(f"\nIdentified numerical columns in Victims DataFrame: {victims_numerical_cols}")
for col_name in victims_numerical_cols:
    detect_outliers_iqr(victims_df, col_name, "Victims")
```

```python
#Remove Outliers

from pyspark.sql.functions import col, lit, when

def handle_outliers_iqr_capping(df, column_name, df_name):
    print(f"\n--- Handling outliers in '{column_name}' of {df_name} DataFrame using capping ---")

    quantiles = df.approxQuantile(column_name, [0.25, 0.75], 0.01)
    q1 = quantiles[0]
    q3 = quantiles[1]
    iqr = q3 - q1

    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    initial_count = df.count()

    # Apply capping logic
    df_capped = df.withColumn(
        column_name,
        when(col(column_name) < lower_bound, lower_bound)
        .when(col(column_name) > upper_bound, upper_bound)
        .otherwise(col(column_name))
    )

    # Count values that were outside the bounds before capping
    outliers_count_before = df.filter((col(column_name) < lower_bound) | (col(column_name) > upper_bound)).count()

    print(f"  Q1: {q1}, Q3: {q3}, IQR: {iqr}")
    print(f"  Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
    print(f"  Number of values capped in '{column_name}': {outliers_count_before} ({outliers_count_before / initial_count:.2%} of initial values)")

    return df_capped


collisions_cols_to_remove_outliers = [
    "distance", "injured_victims", "party_count", "severe_injury_count",
    "other_visible_injury_count", "complaint_of_pain_injury_count",
    "pedestrian_injured_count", "bicyclist_injured_count", "motorcyclist_injured_count",
    "chp_road_type", "pedestrian_collision", "bicycle_collision", "motorcycle_collision", "truck_collision"
]

parties_cols_to_remove_outliers = [
    "party_age", "vehicle_year", "party_number_killed", "party_number_injured"
]

victims_cols_to_remove_outliers = ["victim_age"]


for col_name in collisions_cols_to_remove_outliers:
    collisions_df = handle_outliers_iqr_capping(collisions_df, col_name, "Collisions")


for col_name in parties_cols_to_remove_outliers:
    parties_df = handle_outliers_iqr_capping(parties_df, col_name, "Parties")


for col_name in victims_cols_to_remove_outliers:
    victims_df = handle_outliers_iqr_capping(victims_df, col_name, "Victims")
```

```python
# Query: Find the month with the highest number of collisions

collisions_df.createOrReplaceTempView("collisions_table_month")

# PySpark SQL query to find the month with the highest number of collisions
month_with_highest_collisions_sql = spark.sql("""
    SELECT
        MONTH(collision_date) AS collision_month,
        COUNT(*) AS collision_count
    FROM
        collisions_table_month
    WHERE collision_date IS NOT NULL
    GROUP BY
        collision_month
    ORDER BY
        collision_count DESC
    LIMIT 1
""")

# Convert Spark DataFrame to Pandas for easier mapping
month_with_highest_collisions_pandas = month_with_highest_collisions_sql.toPandas()

# Map month number to month name for better readability
month_names = {
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
}

if not month_with_highest_collisions_pandas.empty:
    month_num = int(month_with_highest_collisions_pandas['collision_month'].iloc[0])
    collision_count = int(month_with_highest_collisions_pandas['collision_count'].iloc[0])
    month_name = month_names.get(month_num, 'Unknown')

    print("Month with the highest number of collisions:")
    print(f"Month: {month_name} ({month_num})")
    print(f"Number of Collisions: {collision_count}")
else:
    print("Could not identify the month with the highest number of collisions.")
```

##**4.5. Fatal Collisions.** <font color = red>[5 marks]</font> <br>

Q. Calculate the percentage of collisions that resulted in fatalities.

```python
# Query: Determine the percentage of collisions that resulted in fatalities

collisions_df.createOrReplaceTempView("fatal_collisions_table")

# PySpark SQL query to calculate the percentage of fatal collisions
fatal_collisions_percentage_sql = spark.sql("""
    SELECT
        CAST(SUM(CASE WHEN killed_victims > 0 THEN 1 ELSE 0 END) AS DOUBLE) / COUNT(*) * 100 AS fatal_collision_percentage
    FROM
        fatal_collisions_table
""")

# Collect the result
fatal_percentage = fatal_collisions_percentage_sql.collect()[0]['fatal_collision_percentage']

print(f"Percentage of collisions that resulted in fatalities: {fatal_percentage:.2f}%")
```
