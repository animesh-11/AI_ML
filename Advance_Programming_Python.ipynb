Write a function ‘longest_unique_substring(s)’ that takes a string ‘s’ as input and returns the length of the longest substring within 's' that contains no repeating characters. A substring is a contiguous sequence of characters within a string.



Input Format:

A single string s.



Output Format:

An integer representing the length of the longest substring without repeating characters.



Constraints:

String 's' must be non-empty and contain only alphabetic characters.
 1 <= len(s) <= 1000




Sample Input:

abcabcbb



Sample Output:

3

````python
def longest_unique_substring(s):
    """
    Finds the length of the longest substring without repeating characters.

    Args:
        s (str): The input string.

    Returns:
        int: The length of the longest substring without repeating characters.
    """
    if not s:
        return 0

    max_length = 0
    start = 0
    char_set = set()

    for end in range(len(s)):
        while s[end] in char_set:
            char_set.remove(s[start])
            start += 1
        char_set.add(s[end])
        max_length = max(max_length, end - start + 1)

    return max_length

# Sample Input:
sample_input = "abcabcbb"
sample_output = longest_unique_substring(sample_input)
print(f"Input: {sample_input}")
print(f"Output: {sample_output}") # Expected Output: 3
````

Write a function ‘three_sum(nums)’ that finds all unique triplets in the list ‘nums’ that sum to zero. The solution set must not contain duplicate triplets.



Input Format:

A list of integers 'nums'.



Output Format:

A list of lists, where each list contains three integers such that their sum is zero.



Constraints:

The input list ‘nums’ will contain at least three integers.
The integers in the list may be both positive and negative.




Sample Input:

[-1, 0, 1, 2, -1, -4]



Sample Output:

[[-1, -1, 2], [-1, 0, 1]]

````python
def three_sum(nums):
    """
    Finds all unique triplets in the list `nums` that sum to zero.

    Args:
        nums (list): A list of integers.

    Returns:
        list: A list of lists, where each inner list contains three integers that sum to zero.
    """
    nums.sort()
    result = []
    n = len(nums)

    for i in range(n - 2):
        # Skip duplicate values for the first element
        if i > 0 and nums[i] == nums[i - 1]:
            continue

        left, right = i + 1, n - 1
        while left < right:
            current_sum = nums[i] + nums[left] + nums[right]

            if current_sum == 0:
                result.append([nums[i], nums[left], nums[right]])
                # Skip duplicate values for the second and third elements
                while left < right and nums[left] == nums[left + 1]:
                    left += 1
                while left < right and nums[right] == nums[right - 1]:
                    right -= 1
                left += 1
                right -= 1
            elif current_sum < 0:
                left += 1
            else:
                right -= 1
    return result

# Sample Input:
sample_input = [-1, 0, 1, 2, -1, -4]
sample_output = three_sum(sample_input)
print(f"Input: {sample_input}")
print(f"Output: {sample_output}") # Expected Output: [[-1, -1, 2], [-1, 0, 1]]
````

You are given three lists representing the scores of students in three subjects: Maths, Science, and English. Each index across the three lists corresponds to a single student.



Write a function evaluate_top_subjects() that:

Uses zip() to iterate over the scores of each student
Determines the subject(s) in which each student scored the highest
Returns a list of strings indicating the top-scoring subject(s) for each student
If there’s a tie between two or more subjects, return all of them in the order: Maths, Science, English, separated by commas.


Input Format:

Three lists of integers: maths, science, and english

Each list contains scores for the same number of students in the respective subjects



Output Format:

A list of strings, where each string represents the top-scoring subject(s) for one student



Constraints:

No upper limit on number of students or scores



Sample Input:

[80, 90, 70, 85]

[85, 90, 95, 80]

[75, 90, 80, 85]



Sample Output:

['Science', 'Maths,Science,English', 'Science', 'Maths,English']

````python
def evaluate_top_subjects(maths, science, english):
    """
    Evaluates the top-scoring subject(s) for each student.

    Args:
        maths (list): List of Maths scores.
        science (list): List of Science scores.
        english (list): List of English scores.

    Returns:
        list: A list of strings, where each string represents the
              top-scoring subject(s) for one student.
    """
    top_subjects_per_student = []

    for m_score, s_score, e_score in zip(maths, science, english):
        current_scores = {'Maths': m_score, 'Science': s_score, 'English': e_score}
        max_score = max(current_scores.values())

        top_subjects = []
        if current_scores['Maths'] == max_score:
            top_subjects.append('Maths')
        if current_scores['Science'] == max_score:
            top_subjects.append('Science')
        if current_scores['English'] == max_score:
            top_subjects.append('English')

        top_subjects_per_student.append(','.join(top_subjects))

    return top_subjects_per_student

# Sample Input:
maths_scores = [80, 90, 70, 85]
science_scores = [85, 90, 95, 80]
english_scores = [75, 90, 80, 85]

sample_output = evaluate_top_subjects(maths_scores, science_scores, english_scores)
print(f"Input Maths: {maths_scores}")
print(f"Input Science: {science_scores}")
print(f"Input English: {english_scores}")
print(f"Output: {sample_output}") # Expected Output: ['Science', 'Maths,Science,English', 'Science', 'Maths,English']
````

Hint: Use enumerate() to track both index and value.



Input Format:

A list of strings called items



Output Format:

A list of strings indicating repeat occurrences and first index

The order of the output should match the order of repeated elements in the input



Constraints:

Input list will not be empty
Elements are case-sensitive
There is no upper limit on the number of elements


Sample Input:

["apple", "banana", "apple", "orange", "banana", "apple", "Apple"]



Sample Output:

['apple repeats at index 2, first found at index 0', 'banana repeats at index 4, first found at index 1', 'apple repeats at index 5, first found at index 0']



Explanation:

1. 'apple' is found to repeat at index 2, first appears at index 0

2. 'banana' is found to repeat at index 4, first appears at index 1

3. 'apple' is again found to repeat at index 5, but appears first at index 0

4. 'Apple' and 'orange' are not found to repeat

````python
def detect_repeats(items):
    """
    Tracks repeated elements in a list and returns a list of formatted strings
    describing the index of the repeated element and its first occurrence.

    Args:
        items (list): A list of strings.

    Returns:
        list: A list of strings indicating repeat occurrences and first index.
    """
    first_occurrence = {}
    result = []

    for i, element in enumerate(items):
        if element in first_occurrence:
            # Element has appeared before, so it's a repeat
            first_idx = first_occurrence[element]
            result.append(f"{element} repeats at index {i}, first found at index {first_idx}")
        else:            # First occurrence of this element
            first_occurrence[element] = i

    return result

# Sample Input:
sample_input = ["apple", "banana", "apple", "orange", "banana", "apple", "Apple"]
sample_output = detect_repeats(sample_input)
print(f"Input: {sample_input}")
print(f"Output: {sample_output}")
# Expected Output: ['apple repeats at index 2, first found at index 0', 'banana repeats at index 4, first found at index 1', 'apple repeats at index 5, first found at index 0']
````

- The operation must be one of 'sum', 'product', or 'average'

- All numbers must be valid floats or integers



Sample Input:

````python
import math

def operate_on_numbers(operation_type, *args):
    """
    Performs various mathematical operations on a variable number of input values.

    Args:
        operation_type (str): The type of operation ('sum', 'product', or 'average').
        *args: A variable number of numerical arguments.

    Returns:
        float or str: The result of the operation, or an error message.
    """
    if operation_type not in ['sum', 'product', 'average']:
        return "Invalid operation specified"

    if not args:
        return "No numbers provided"

    numbers = [float(arg) for arg in args]

    if operation_type == 'sum':
        return sum(numbers)
    elif operation_type == 'product':
        return math.prod(numbers)
    elif operation_type == 'average':
        return sum(numbers) / len(numbers)

# Sample Input:
# The input format implies parsing a string. Let's demonstrate by calling the function directly.
# 'sum 1 2 3 4'

# To simulate the input format, we can parse it:
sample_input_str = "sum 1 2 3 4"
parts = sample_input_str.split()
operation = parts[0]
nums_str = parts[1:]
nums = [float(n) for n in nums_str]

sample_output = operate_on_numbers(operation, *nums)
print(f"Operation: {operation}, Numbers: {nums_str}")
print(f"Result: {sample_output}")

# Example for product:
sample_output_product = operate_on_numbers('product', 2, 3, 4)
print(f"\nOperation: product, Numbers: [2, 3, 4]")
print(f"Result: {sample_output_product}")

# Example for average:
sample_output_average = operate_on_numbers('average', 10, 20, 30)
print(f"\nOperation: average, Numbers: [10, 20, 30]")
print(f"Result: {sample_output_average}")

# Example for no numbers:
sample_output_no_nums = operate_on_numbers('sum')
print(f"\nOperation: sum, Numbers: []")
print(f"Result: {sample_output_no_nums}")

# Example for invalid operation:
sample_output_invalid = operate_on_numbers('subtract', 5, 2)
print(f"\nOperation: subtract, Numbers: [5, 2]")
print(f"Result: {sample_output_invalid}")
````

You are given a list of nested tuples, where each tuple contains exactly two integers.



Write a function transform_tuples() that applies the following transformations to each tuple using map():

If both numbers in the tuple are even, return their sum
If both numbers are odd, return their product
If one is even and one is odd, return their absolute difference
Return a list of results after applying the above logic to all tuples in the input list.


Input Format:

A list of tuples, where each tuple contains exactly two integers



Output Format:

A list of integers after applying the defined transformation logic to each tuple



Constraints:

Assume 1 ≤ length of list
Assume each number in a tuple is a positive integer
No upper bound is imposed on the number values or list length


Sample Input:

[(2, 4), (3, 5), (2, 5), (8, 6), (7, 2)]



Sample Output:

[6, 15, 3, 14, 5]

````python
def transform_tuples(list_of_tuples):
    """
    Applies specific transformations to each tuple based on the parity of its elements.

    Args:
        list_of_tuples (list): A list of tuples, where each tuple contains two integers.

    Returns:
        list: A list of integers after applying the defined transformation logic.
    """

    def apply_transformation(t):
        num1, num2 = t
        is_num1_even = num1 % 2 == 0
        is_num2_even = num2 % 2 == 0

        if is_num1_even and is_num2_even:
            return num1 + num2  # Both even, return sum
        elif not is_num1_even and not is_num2_even:
            return num1 * num2  # Both odd, return product
        else:
            return abs(num1 - num2)  # One even, one odd, return absolute difference

    return list(map(apply_transformation, list_of_tuples))

# Sample Input:
sample_input = [(2, 4), (3, 5), (2, 5), (8, 6), (7, 2)]
sample_output = transform_tuples(sample_input)
print(f"Input: {sample_input}")
print(f"Output: {sample_output}") # Expected Output: [6, 15, 3, 14, 5]
````

A palindrome is a word that reads the same forwards and backwards, ignoring case. For example, "Madam", "racecar" and "Deified" are all palindromes.



Your task is to write a Python function filter_palindromes(word_list) that returns only the palindrome words from a list of words. The function must:

Use the built-in filter() function along with a lambda expression
Compare each word with its reverse (after converting both to lowercase)
Return a list of all such words, preserving the original casing


Input Format:

A list of strings (words)



Output Format:

A list of strings containing only the palindromes from the input list



Constraints:

1 ≤ number of words ≤ 100
1 ≤ length of each word ≤ 50
Words may be in any case
The input list may be empty


Sample Input:

['racecar', 'hello', 'Madam', 'world', 'civic']



Sample Output:

['racecar', 'Madam', 'civic']

````python
def filter_palindromes(word_list):
    """
    Filters a list of words to return only palindromes, ignoring case.

    Args:
        word_list (list): A list of strings.

    Returns:
        list: A list of strings containing only the palindromes,
              preserving their original casing.
    """
    # Lambda function to check if a word is a palindrome (case-insensitive)
    is_palindrome = lambda word: word.lower() == word.lower()[::-1]

    # Use filter() with the lambda function to get the palindromes
    return list(filter(is_palindrome, word_list))

# Sample Input:
sample_input = ['racecar', 'hello', 'Madam', 'world', 'civic']
sample_output = filter_palindromes(sample_input)
print(f"Input: {sample_input}")
print(f"Output: {sample_output}") # Expected Output: ['racecar', 'Madam', 'civic']
````

Returns a list of tuples (x, y) for each corresponding element from lists a and b
Filters out any pair where either x or y is None
Use a list comprehension with zip() to implement this.


Input Format:

A list of elements a
A list of elements b
Both lists are taken one after another and are of the same length.


Output Format:

A list of tuples after filtering out any with None values



Constraints:

len(a) == len(b)
Elements may be of any type, including None


Sample Input:

[1, 2, None]

['x', 'y', 'z']



Sample Output:

[(1, 'x'), (2, 'y')]

You are building a small utility to safely extract an integer from a string. This function must:



Try to convert the given string to an integer
If the conversion fails, return "Invalid input"
Always print "Attempted conversion" using a finally block, regardless of whether the conversion was successful or not.


This is a common pattern in logging and debugging scenarios where operations need to be attempted safely with appropriate messaging.



Write a function extract_number(s) that follows the behaviour described above.



Input Format:

A single string



Output Format:

Always prints "Attempted conversion"
Returns either:
The converted integer if valid, or
The string "Invalid input" if the conversion fails

Sample Output 1:

Attempted conversion

123




Sample Input 2:

hello



Sample Output 2:

Attempted conversion

Invalid input





Sample Input 3:

3.14



Sample Output 3:

Attempted conversion

````python
def extract_number(s):
    """
    Attempts to convert a string to an integer, handling invalid inputs.
    Always prints 'Attempted conversion' using a finally block.

    Args:
        s (str): The input string.

    Returns:
        int or str: The converted integer if valid, or 'Invalid input' if conversion fails.
    """
    try:
        return int(s)
    except ValueError:
        return "Invalid input"
    finally:
        print("Attempted conversion")

# Sample Input 1:
print(f"\nSample Input: '123'")
output1 = extract_number("123")
print(output1)

# Sample Input 2:
print(f"\nSample Input: 'hello'")
output2 = extract_number("hello")
print(output2)

# Sample Input 3:
print(f"\nSample Input: '3.14'")
output3 = extract_number("3.14")
print(output3)
````
