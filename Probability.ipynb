To solve this, you need to write a Python function probability_of_pass(selected_students) that takes a list of dictionaries as input. Each dictionary contains three keys, corresponding to the subjects, with boolean values indicating whether the student has passed or failed in each respective subject. The function should calculate the probability that a randomly selected student has passed at least one of the subjects (Mathematics, Science, or English).


Input format


A list of dictionaries representing the performance of students. Each dictionary contains:
Three keys: 'Mathematics', 'Science', and 'English'
The values (bool) for these keys, where True indicates a pass and False indicates a fail in the respective subject

Output format


The probability (float) that a randomly selected student has passed at least one subject

Constraints


The input list contains at least one student
The computed probability is rounded to four decimals

Example Case 1

Input

[{'Mathematics': True, 'Science': False, 'English': False}]


Output

1.0


Example Case 2

Input

[{'Mathematics': False, 'Science': False, 'English': False}]


Output

0.0

```python
def probability_of_pass(selected_students):
    """
    Calculates the probability that a randomly selected student has passed at least one subject.

    Args:
        selected_students: A list of dictionaries representing the performance of students.
                           Each dictionary contains keys 'Mathematics', 'Science', and 'English'
                           with boolean values indicating pass (True) or fail (False).

    Returns:
        The probability (float) that a randomly selected student has passed at least one subject,
        rounded to four decimal places.
    """
    num_students = len(selected_students)
    if num_students == 0:
        return 0.0

    passed_at_least_one = 0
    for student in selected_students:
        if student['Mathematics'] or student['Science'] or student['English']:
            passed_at_least_one += 1

    probability = passed_at_least_one / num_students
    return round(probability, 4)

# Example Usage (from the problem description):
print(probability_of_pass([{'Mathematics': True, 'Science': False, 'English': False}]))
print(probability_of_pass([{'Mathematics': True, 'Science': True, 'English': True}, {'Mathematics': False, 'Science': True, 'English': False}, {'Mathematics': False, 'Science': False, 'English': True}]))
```

In a game simulation involving a die with n sides, you need to calculate the probability of achieving a specific sum after rolling the die a certain number of times. For example, if you roll a 6-sided die twice, you may want to calculate the probability of obtaining a sum of 7.


To solve this, you need to write a Python function probability_of_outcome(num_sides, rolls, target_sum) that takes three parameters: the number of sides on the die, the number of times the die will be rolled, and the target sum that you want to calculate the probability for across the rolls.


The function should return the probability of obtaining the target sum after rolling the die the specified number of times


Input format


A tuple whose elements, respectively, are

num_sides: the number (int) of sides on the die
rolls: the number (int) of die rolls
target_sum: the target sum (int) to calculate the probability for

Output format


A floating-point number (float) representing the probability of rolling the target sum with the given number of rolls

Constraints


The face in contact of the floor should be considered to contain the value to be summed (this is to avoid cases where there isn't a face pointing upwards, such as in the case of a 4-sided die)
Each side of the die has an equal probability of landing
The sides are each numbered in sequence (e.g., a 3 sided die will be numbered 1, 2, and 3 on its respective sides)
num_sides >= 4 to avoid the situation of impossible die structures in 3D
1 <= rolls <= 6
The computed probability is rounded to four decimals

Example Case 1

Input

(6, 2, 7)


Output

0.1667


Example Case 2

Input

(4, 4, 4)


Output

0.0039

```python
def probability_of_outcome(num_sides, rolls, target_sum):
    """
    Calculates the probability of achieving a specific sum after rolling a die
    with a given number of sides a certain number of times.

    Args:
        num_sides: The number of sides on the die (int).
        rolls: The number of times the die will be rolled (int).
        target_sum: The target sum to calculate the probability for (int).

    Returns:
        The probability (float) of rolling the target sum, rounded to four decimal places.
    """
    # Dynamic programming approach to count the number of ways to get the target sum
    # dp[i][j] will store the number of ways to get sum j with i rolls
    dp = [[0 for _ in range(target_sum + 1)] for _ in range(rolls + 1)]
    dp[0][0] = 1

    for i in range(1, rolls + 1):
        for j in range(1, target_sum + 1):
            for k in range(1, num_sides + 1):
                if j - k >= 0:
                    dp[i][j] += dp[i - 1][j - k]

    # Total possible outcomes is num_sides ^ rolls
    total_outcomes = num_sides ** rolls

    # The number of ways to get the target sum
    ways_to_target_sum = dp[rolls][target_sum]

    # Calculate the probability and round to four decimal places
    probability = ways_to_target_sum / total_outcomes
    return round(probability, 4)

# Example Usage (from the problem description):
print(probability_of_outcome(6, 2, 7))
print(probability_of_outcome(4, 4, 4))
```

You are a quality assurance analyst working in a factory that manufactures electronic components, such as chips or connectors. Each day, the factory produces thousands of units, and your job is to monitor the defect rate to ensure quality standards are met.


At the end of the production run, a random sample is tested, and the number of defective units is recorded. Your task is to calculate the probability that a randomly selected item from the entire batch is not defective.


Input Format


A tuple containing, in sequence

An integer n representing the size of the sample
An integer d representing the number of defective units in the sample

Output Format


A float value representing the probability that a randomly selected unit is not defective

Constraints


n > 0
d >= 0
n >= d
The sample accurately represents the batch
The computed probability is rounded to 4 decimal places

Example Case 1


Input


(100, 27)


Output


0.73


Example Case 2


Input

(6, 2)

```python
def probability_of_non_defective(sample_size, defective_units):
    """
    Calculates the probability that a randomly selected item from a batch is not defective.

    Args:
        sample_size: The size of the sample (int).
        defective_units: The number of defective units in the sample (int).

    Returns:
        The probability (float) that a randomly selected unit is not defective,
        rounded to four decimal places.
    """
    non_defective_units = sample_size - defective_units
    probability = non_defective_units / sample_size
    return round(probability, 4)

# Example Usage (from the problem description):
print(probability_of_non_defective(100, 27))
print(probability_of_non_defective(6, 2))
```

You work as a marketing analyst for an e-commerce company. Your team recently launched two promotional campaigns:


Campaign A: a discount code for mobile users
Campaign B: a coupon for newsletter subscribers

Customers may qualify for either campaign or both. In this case, you're told that every customer is qualified for at least one campaign. Your task is to calculate the probability that a randomly selected customer qualified for both Campaign A and Campaign B.


Write a function campaign_overlap_probability(p_A, p_B) that accepts the probability of a customer qualifying for Campaign A and for Campaign B and returns the probability that a customer qualified for both campaigns


Input Format


A tuple containing the following, in sequence

p_A (float): the probability of qualifying for Campaign A
p_B (float): the probability of qualifying for Campaign B

Output Format


A floating-point number (float) representing the probability that a customer qualified for both campaigns

Constraints


0 ≤ p_A ≤ 1
0 ≤ p_B ≤ 1
The computed probability is rounded to two decimals

Example Case 1

Input

(0.6531, 0.5872)


Output

0.24

Example Case 2

Input

(0.72345, 0.68987)

Output

0.41

```python
def campaign_overlap_probability(p_A, p_B):
    """
    Calculates the probability that a customer qualified for both Campaign A and Campaign B.

    Args:
        p_A: The probability of qualifying for Campaign A (float).
        p_B: The probability of qualifying for Campaign B (float).

    Returns:
        The probability (float) that a customer qualified for both campaigns,
        rounded to two decimal places.
    """
    # Since every customer is qualified for at least one campaign, P(A U B) = 1
    # Using the formula P(A U B) = P(A) + P(B) - P(A ∩ B)
    # We can rearrange to find P(A ∩ B) = P(A) + P(B) - P(A U B)
    probability_overlap = p_A + p_B - 1.0
    return round(probability_overlap, 2)

# Example Usage (from the problem description):
print(campaign_overlap_probability(0.6531, 0.5872))
print(campaign_overlap_probability(0.72345, 0.68987))
```

In a game of Snakes and Ladders, you are given a board with numbered squares. You are  squares away from the finish line, where  can be any non-negative integer. The goal is to land exactly on the final square.


You roll a 6-sided die and can move between 1 to 6 squares per roll. If you exceed the target square, you go back the remaining steps from the end.


For example, if you are 2 squares away from the end (given by X), and you roll 3, you will end up 1 square away from the end

i.e., the path would be (X-2) -> (X-1) -> X -> (X-1)

You are tasked with calculating the probability that you will land exactly on the final square in a given number of rolls, k.


Write a Python function probability_of_landing_on_final_square(n, k) that calculates the probability of landing exactly on the final square after a set number of rolls in a game. The function should take two inputs: the number of squares remaining before reaching the final square and the number of dice rolls. The output should be the probability that, after rolling the dice the specified number of times, you will land exactly on the final square.


Input format

A tuple containing, sequentially
n (int): represents how many squares are left before the final square
k (int): represents the number of rolls

Output format

A floating-point number (float) representing the probability that you will land exactly on the final square in k rolls

Constraints

n ≥ 1
k ≥ 1
The computed probability is rounded to four decimals

Example Case 1

Input

(4, 1)


Output

0.1667


Example Case  2

Input

(50, 1)


Output

0.0

```python
def probability_of_landing_on_final_square(n, k):
    """
    Calculates the probability of landing exactly on the final square after k rolls
    in a game of Snakes and Ladders with a special rule for overshooting.

    Args:
        n: The number of squares remaining before the final square (int).
        k: The number of rolls (int).

    Returns:
        The probability (float) of landing exactly on the final square in k rolls,
        rounded to four decimal places.
    """
    # dp[i][j] will store the probability of being j squares away from the finish line after i rolls
    dp = [[0.0 for _ in range(n + 6)] for _ in range(k + 1)]

    # Base case: After 0 rolls, the probability of being n squares away is 1.0
    dp[0][n] = 1.0

    for i in range(1, k + 1):
        for j in range(n + 6):  # Iterate through possible distances from the finish line
            if dp[i-1][j] > 0:  # If there's a non-zero probability of being at this distance
                for roll in range(1, 7):  # Iterate through possible die rolls (1 to 6)
                    next_pos = j - roll
                    if next_pos < 0:
                        # If overshooting, the new position is the absolute value of the overshoot
                        next_pos = abs(next_pos)
                    dp[i][next_pos] += dp[i-1][j] * (1/6.0)

    # The probability of landing exactly on the final square (distance 0) after k rolls
    return round(dp[k][0], 4)

# Example Usage (from the problem description):
print(probability_of_landing_on_final_square(4, 1))
print(probability_of_landing_on_final_square(50, 1))
```

You are working as a data analyst for an e-commerce platform that collects feedback on product preferences. The platform shows users a variety of products and asks them to rate their likelihood of liking each one.


Based on historical data, the probability that a customer likes a particular product is estimated, and the results for each product are stored in a dictionary. The probabilities for each product are independent of each other.


Your task is to write a function calculate_joint_probability(products_and_probabilities) that returns the joint probability that a customer likes all the products


Input Format


A dictionary where:


The keys are product names (str)
The values are the probabilities (floats) representing the likelihood that a customer likes each product
The dictionary must contain at least one product (i.e., it cannot be empty)

Output Format


The joint probability (float) that the customer likes all the products

Constraints


The computed probability is rounded to 4 decimal places

{'headphones': 0.7, 'backpack': 0.5, 'laptop': 0.8}

```python
def calculate_joint_probability(products_and_probabilities):
    """
    Calculates the joint probability that a customer likes all the products.

    Args:
        products_and_probabilities: A dictionary where keys are product names (str)
                                    and values are the probabilities (floats) of liking each product.

    Returns:
        The joint probability (float) that the customer likes all the products,
        rounded to four decimal places.
    """
    joint_probability = 1.0
    for probability in products_and_probabilities.values():
        joint_probability *= probability

    return round(joint_probability, 4)

# Example Usage (from the problem description):
# print(calculate_joint_probability({'headphones': 0.7, 'backpack': 0.5, 'laptop': 0.8}))
# print(calculate_joint_probability({'headphones': 1.0, 'backpack': 1.0, 'laptop': 1.0}))
```

As a risk analyst at an insurance company, you are tasked with evaluating the likelihood of various claims for properties, including fire, flood, and theft. Based on historical data, you are provided with the probabilities of each individual claim, as well as the likelihood of multiple claims occurring simultaneously, either in pairs or all three together.


Your task is to calculate the probability that a randomly selected policyholder will make at least one type of claim


Write a function claim_risk_exposure(p_fire, p_flood, p_theft, p_fire_flood, p_fire_theft, p_flood_theft, p_all_three) that accepts the probabilities as inputs and returns the probability of at least one claim


Input Format


A tuple containing, in sequence
p_fire (float): Probability of fire claim
p_flood (float): Probability of flood claim
p_theft (float): Probability of theft claim
p_fire_flood (float): Probability of both fire and flood
p_fire_theft (float): Probability of both fire and theft
p_flood_theft (float): Probability of both flood and theft

Output Format


The probability (float) that a property has at least one claim

Constraints

0 ≤ p_flood ≤ 1
0 ≤ p_theft ≤ 1

0 ≤ p_all_three ≤ 1

(0.45, 0.2, 0.35, 0.15, 0.12, 0.09, 0.05)

```python
def claim_risk_exposure(p_fire, p_flood, p_theft, p_fire_flood, p_fire_theft, p_flood_theft, p_all_three):
    """
    Calculates the probability that a randomly selected policyholder will make at least one type of claim.

    Args:
        p_fire: Probability of fire claim (float).
        p_flood: Probability of flood claim (float).
        p_theft: Probability of theft claim (float).
        p_fire_flood: Probability of both fire and flood (float).
        p_fire_theft: Probability of both fire and theft (float).
        p_flood_theft: Probability of both flood and theft (float).
        p_all_three: Probability of all three events occurring together (float).

    Returns:
        The probability (float) that a property has at least one claim,
        rounded to two decimal places.
    """
    # Using the principle of inclusion-exclusion for three events:
    # P(A U B U C) = P(A) + P(B) + P(C) - P(A ∩ B) - P(A ∩ C) - P(B ∩ C) + P(A ∩ B ∩ C)
    probability_at_least_one_claim = (p_fire + p_flood + p_theft -
                                      p_fire_flood - p_fire_theft - p_flood_theft +
                                      p_all_three)

    return round(probability_at_least_one_claim, 2)

# Example Usage (from the problem description):
# print(claim_risk_exposure(0.3, 0.25, 0.2, 0.1, 0.08, 0.05, 0.03))
# print(claim_risk_exposure(0.45, 0.2, 0.35, 0.15, 0.12, 0.09, 0.05))
```

You work as a junior data analyst in a city's weather forecasting department. The meteorological model outputs the probability of rain for each day based on atmospheric conditions, humidity, wind patterns, and historical data.


For every day in the near future, the system forecasts a 30% chance of rain. Write a Python function calculate_rain_probability_in_fortnight(n) that computes the probability that it will rain on exactly n days out of the next 14 days, assuming the events are independent.


Input Format


n: The number (int) of days (out of 14) that you want to calculate the probability of rainfall for

Output Format


The probability (float) that it will rain on exactly n days out of the next 14 days

Constraints


0 ≤ n ≤ 14
The computed probability is rounded to four decimal places

Example Case 1

Input

3


Output

0.1943


Example Case 2

Input

5


Output

0.1963

```python
import math

def calculate_rain_probability_in_fortnight(n):
    """
    Computes the probability that it will rain on exactly n days out of the next 14 days,
    assuming the events are independent.

    Args:
        n: The number (int) of days (out of 14) that you want to calculate the probability of rainfall for.

    Returns:
        The probability (float) that it will rain on exactly n days out of the next 14 days,
        rounded to four decimal places.
    """
    # Total number of days
    N = 14
    # Probability of rain on a single day
    p = 0.30
    # Probability of no rain on a single day
    q = 1 - p

    # Binomial probability formula: P(X=n) = C(N, n) * p^n * (1-p)^(N-n)
    # C(N, n) = N! / (n! * (N-n)!) - Binomial coefficient
    if n < 0 or n > N:
        return 0.0

    combinations = math.comb(N, n)
    probability = combinations * (p ** n) * (q ** (N - n))

    return round(probability, 4)

# Example Usage (from the problem description):
print(calculate_rain_probability_in_fortnight(3))
print(calculate_rain_probability_in_fortnight(5))
```
